```python
#Ignorar os erros
import warnings
warnings.filterwarnings('ignore')
#pandas
import pandas as pd 
```

```python
#importando os dataset's

df_agregados = pd.read_csv('./AGREGADOS_02.csv',delimiter=';')

df_pacotes = pd.read_csv('./pacotes_entregues_310124_processado.csv',delimiter=',')
```

```python
df_valores = pd.read_csv('./',delimiter=';')
```

```python
df_valores.head(2)
```

```python
df_pacotes.head()
```

```python
#Fazendo o Join  entre o dataframe df_pacotes X df_agregados

df_pacotes_agregados = pd.merge(df_pacotes,df_agregados,on=['CPF_CNPJ'],how='left')
                                
df_pacotes_agregados.shape
```

```python
#Exibindo o dataframe apos o Join

df_pacotes_agregados.head()
```

```python
#Excluindo a colunas que replicou

df_pacotes_agregados.drop(['AGREGADO_y'],axis=1,inplace=True)

#Modificando o nome da colunas

df_pacotes_agregados.rename(columns={'AGREGADO_x':'AGREGADO'},inplace=True)

#Exibindo as info 

df_pacotes_agregados.info()
```

```python
#Exibindi informação do dataframe 

df_valores.info()
```

```python
#Realizando um Join com o dataframe de valores

df_pacotes_full = pd.merge(df_pacotes_agregados,df_valores,on=['CIDADE','BAIRRO','TIPO_VEICULO'],how='left')

#Exibindo a dimensão

df_pacotes_full.shape
```

```python
#Verificando os nulos e quantidade a mais de pacotes

df_pacotes_full.isnull().sum()
```

```python
#Filtrando os pacotes nulos

df_nulos = df_pacotes_full[df_pacotes_full['VALOR'].isna()]

df_nulos.shape
```

```python
#Exibindo o dataframe

df_nulos.isnull().sum()
```

```python
#Excluíndo os atributos para realizar um novo join

df_nulos.drop(['VALOR','BASE','MARCA_VEICULO','BASE_VINCULADO','PLACA'],axis=1,inplace=True)

#Exibindo os atributos

df_nulos.info()
```

```python
#Realizando o JOin com os pedidos que não possuem bairros cadastrados

df_nulos_proc = pd.merge(df_nulos,df_valores,on=['CIDADE','TIPO_VEICULO'],how='left')

#Exibindo a dimensão

df_nulos_proc.shape
```

```python
#Excluí os duplicados

df_nulos_proc.drop_duplicates(subset=['NUMERO_PEDIDO_JMS'],inplace=True)

df_nulos_proc.shape
```

```python
#Exibindo o dataframe

df_nulos_proc.head()
```

```python
#Excluíndo o atributo duplicado

df_nulos_proc.drop(['BAIRRO_y'],axis=1,inplace=True)

#modificando o nome do atributo

df_nulos_proc.rename(columns={'BAIRRO_x':'BAIRRO'},inplace=True)

#exibindo as info

df_nulos_proc.info()
```

```python
df_nulos_proc.isnull().sum()
```

```python
#verificandos o valores que ainda costam como nulo

df_nulos_proc02 = df_nulos_proc[df_nulos_proc['VALOR'].isna()]

df_nulos_proc02.head()
```

Excluído os missings do daframe dos pacotes para depois fazer a concatenação

```python
df_pacotes_full.dropna(subset=['VALOR'],inplace=True)

df_pacotes_full.isnull().sum()
```

```python
df_nulos_proc.shape
```

```python
#excluído os atributos

df_pacotes_full.drop(['PLACA','MARCA_VEICULO','BASE_VINCULADO'],axis=1,inplace=True)

#Exibindo as info

df_pacotes_full.info()
```

```python
#Concatenando

df_fechamento_full = pd.concat([df_pacotes_full,df_nulos_proc])

df_fechamento_full.shape
```
